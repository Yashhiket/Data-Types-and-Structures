1. What are data structures, and why are they important?

Answer:
Data structures are ways to organize and store data in a computer so it can be accessed and modified efficiently.
They are important because they help in:

Efficient data storage (e.g., arrays, lists)

Fast data retrieval (e.g., dictionaries)

Better code organization and performance.

2. Difference between mutable and immutable data types with examples

Answer:

Mutable: Can be changed after creation (e.g., list, dict, set)

Copy
Edit

Immutable: Cannot be changed after creation (e.g., int, str, tuple)

18. When are lists preferred over dictionaries?

Answer:

When order matters.

19. Why are dictionaries considered unordered, and how does it affect data retrieval?

Answer:
Earlier (before Python 3.7), dictionaries did not maintain insertion order.
Now, they do, but retrieval is still by key, not position.
Access is always key-based, not index-based.
